```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sns
```

```python
df = pd.read_csv("resultados_predicao.csv")
df
```

```python
df_cpu = df[df['metrica'] == 'CPU']
```

```python
df_mem = df[df['metrica'] == 'Memory']
```

```python
df_lat = df[df['metrica'] == 'Latency']
```

```python
df_req = df[df['metrica'] == 'Request']
```

```python
fig, ax = plt.subplots(figsize=(15,5)) 

labels = ["Redes Neurais \nRecorrentes", "Redes Neurais\n Convolucionais", "XGBoost", "AutoArima"]

#Plot 1
plt.subplot(1, 2, 1) 
y1 = df_cpu["r2_score"]

plot1 = sns.barplot(x='modelo',y=y1,data=df_cpu, hue="API") 

plot1.set_xticks(range(4))
plot1.set_xticklabels(labels)

plt.xlabel("Modelo") 
plt.ylabel("Coeficiente de determinaÃ§Ã£o (RÂ²)") 
plt.title("Coeficiente de determinaÃ§Ã£o (RÂ²) da CPU por Modelo e AplicaÃ§Ã£o", loc="center", fontsize=12)
plt.tight_layout() 

#Plot 2
plt.subplot(1, 2, 2) 
y2 = df_cpu["tempo"]

plot2 = sns.barplot(x='modelo',y=y2,data=df_cpu, hue="API") 

plot2.set_xticks(range(4))
plot2.set_xticklabels(labels)

plt.xlabel("Modelo") 
plt.ylabel("Tempo de execuÃ§Ã£o (em segundos)") 
plt.title("Tempo de execuÃ§Ã£o da CPU por Modelo e AplicaÃ§Ã£o", loc="center", fontsize=12) 

plt.tight_layout() 
plt.show()
```
